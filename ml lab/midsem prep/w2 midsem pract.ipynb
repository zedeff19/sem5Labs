```python
#additional
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a sample DataFrame
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Product_A": [100, 120, 130, 110, 140, 150, 160, 170, 180, 190, 200, 210],
    "Product_B": [150, 130, 160, 140, 170, 160, 150, 180, 190, 200, 210, 220],
    "Product_C": [200, 210, 190, 180, 170, 160, 150, 140, 130, 120, 110, 100]
}

df = pd.DataFrame(data)
print(df)
# Step 2: Set the Month column as the index for better plotting
df.set_index('Month', inplace=True)
print(df)
# Step 3: Plot the multiline graph for sales data of each product
plt.figure(figsize=(12, 6))
for column in df.columns:
    plt.plot(df.index, df[column], marker='o', label=column)

plt.title('Monthly Sales Data for Each Product')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend(['a','b','c'],title='Products')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a sample DataFrame (same as above)
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Product_A": [100, 120, 130, 110, 140, 150, 160, 170, 180, 190, 200, 210],
    "Product_B": [150, 130, 160, 140, 170, 160, 150, 180, 190, 200, 210, 220],
    "Product_C": [200, 210, 190, 180, 170, 160, 150, 140, 130, 120, 110, 100]
}

df = pd.DataFrame(data)

# Step 2: Calculate the total units sold for each product over the year
total_sales = df.sum(numeric_only=True)
print(total_sales)
# Step 3: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=total_sales.index, autopct='%2.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Total Sales of Each Product for the Last Year')
plt.show()
```
