```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt

arr = np.linspace(0, 2*pi, 100)
y = np.sin(arr)

plt.plot(arr, y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['sin curve'], title = "curves")
plt.show()
```
